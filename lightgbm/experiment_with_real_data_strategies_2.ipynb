```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
df = pd.read_csv('./../data/airline-passengers.csv', parse_dates=['Month'], index_col='Month')
df.head()
```

```python
# Apply differencing
df_diff = df['Passengers'].diff().fillna(0)  # Using fillna(0) for the first NaN value

# Plot the differenced series
plt.figure(figsize=(14, 7))
plt.plot(df.index, df_diff, label='Differenced Series')
plt.title('Differenced Flight Passenger Series')
plt.legend()
plt.show()
```

#### To reverse trend

```python
# To reverse the differencing and restore the original series
df_restored = df_diff.cumsum() + df['Passengers'].iloc[0]

# Plot the restored series alongside the original
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Passengers'], label='Original Series')
plt.plot(df.index, df_restored, label='Restored Series', linestyle='--')
plt.plot(df.index, df_diff, label='Differenced Series')
plt.title('Original vs Restored Series from Differencing')
plt.legend()
plt.show()
```
